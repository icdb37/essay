## matplotlib

函数简述：  
```py
import matplotlib.pyplot as plt
import numpy as np

fig1,ax1 = plt.subplots()
ax1.plot([0,10],[0,5])  # 两点连线

fig2,ax2 = plt.subplots(nrows=2,ncols=2)
ax2[0][0].plot([-1,0],[0,1])
ax2[0][1].plot([0,1],[1,0])
ax2[1][0].plot([-1,0],[0,-1])
ax2[1][1].plot([0,1],[-1,0])
```
* 参数：
    * `nrows` 行数，默认为1；一行有多少子图
    * `ncols` 列数，默认为1；一列有多少子图
    * `sharex` 共享x轴，默认False
    * `sharey` 共享y轴，默认False
* 返回值：
    * `fig` 表示图形窗口
    * `ax` 子图


````python
import matplotlib.pyplot as plt
import numpy as np

# 图形窗口1
x = np.linspace(0, 2 * np.pi, 20)  # [0,2PI]之间插值共20个点，返回数组
y = np.sin(x)
fig1,ax1 = plt.subplots()
ax1.plot(x,y)
ax1.plot([0,0.2, 0.4, 0.8, 1,6], [0, 0.2, 0.4, 0.8, 1,-1])

# 图形窗口2
fig2,ax2 = plt.subplots(nrows=2,ncols=2)
ax2[0][0].plot([-1,0],[0,1])
ax2[0][1].plot([0,1],[1,0])
ax2[1][0].plot([-1,0],[0,-1])
ax2[1][1].plot([0,1],[-1,0])

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np


def f(a):
    b = a * 2
    while np.linalg.norm(b, ord=2) < 20:
        b = b * 2
    if b.sum() > 10:
        c = b
    else:
        c = 10 * b
    return c

x = np.linspace(-20, 20, 50)
y = f(x)

fig0,ax0 = plt.subplots()
ax0.plot(x,y)

plt.show()
````
